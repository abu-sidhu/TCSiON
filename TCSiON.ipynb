## PREPROCESSING

### i) Removing Biased Features

```python
df['EmployeeCount'].nunique()
```

Employee count is always 'one' for all observations. So, it’s not needed for the machine learning model.

```python
df['EmployeeNumber'].nunique()
```

Employee number is unique for each employee so it shows no relevance in modeling.

```python
df['Over18'].nunique()
```

All employees are over 18 years in the dataset so not needed that column as well

```python
df['StandardHours'].nunique()
```

Standard hours also a common value (80) for all employees so, can be removed that column also

```python
# Dropping unwanted columns 
df=df.drop(['Attrition','EmployeeCount','EmployeeNumber','Over18','StandardHours'],axis=1)
```

Dropping all the columns with no relevance from the dataset. Not checking the attrition rate here so, dropping that column for reducing complications. By keeping previous employees detail, the model can be trained with more observations and hence get better salary prediction. 

### ii) Outlier Detection

```python
# Setting plot size
plt.figure(figsize=(4,6))
plt.subplot(2,1,1)
#Plotting boxplot of Number of companies worked 
sns.boxplot(y=df['NumCompaniesWorked'])
plt.title("Number of Companies Worked")
plt.subplot(2,1,2)
plt.title("Last Year training Time")
#Plotting boxplot of Last Year training Time
sns.boxplot(y=df['TrainingTimesLastYear'])
plt.tight_layout()
plt.show()
```

```python
for i in ['NumCompaniesWorked','TrainingTimesLastYear']:
#Finding the quartile 1 by using quantile
    q1 = df[i].quantile(0.25) 
#Finding the quartile 3 by using quantile
    q3 = df[i].quantile(0.75) 
#     Inter quartile range calculation
    iqr = q3 - q1 
#     Setting Upperlimit and Lower Limit
    upper_limit = q3 + 1.5 * iqr
    lower_limit = q1 - 1.5 * iqr 
#     outlier replacement with upper limit and lower limit instead of removing the observations
    df[i][df[i]>upper_limit] = upper_limit
    df[i][df[i]<lower_limit] = lower_limit
```

```python
#Plotting the same boxplots after outlier handling
plt.figure(figsize=(4,6))
plt.subplot(2,1,1)
sns.boxplot(y=df['NumCompaniesWorked'])
plt.title("Number of Companies Worked")
plt.subplot(2,1,2)
plt.title("Last Year Training Time")
sns.boxplot(y=df['TrainingTimesLastYear'])
plt.tight_layout()
plt.show()
```

### iii) Label Encoding

```python
# Importing label Encoder from sklearn preprocessing libraries
from sklearn.preprocessing import LabelEncoder
label_en=LabelEncoder()
# All the columns in df having object datatype is Label encoded  
for i in df.columns:
    if df[i].dtype == object:
        df[i]=label_en.fit_transform(df[i])
# Checking the data df after label encoding
df.head()
```

### iv) Feature Reduction

```python
# Seperating target
y = df['MonthlyIncome']
x = df.drop('MonthlyIncome', axis = 1)
```

Splitting the dataset to y (target) as  monthly salary and x as all other features

```python
# Correlation matrix of x is created and saved it to variable corr
corr=x.corr()
plt.figure(figsize=(16,16))
# A heat map of correlation matrix with annotation is plotted
ax = sns.heatmap(corr, annot=True, linewidths=0.1, linecolor='white',cmap="twilight_shifted")
```

Since the dataset has limited number of observations, 0.7 and higher correlated features can be removed by keeping one among the correlated features.

'Job level', 'Performance Rating', 'Years in Current Role' and 'Years with current Manager' can be removed by the above condition

```python
x.columns
```

```python
# Dropping highly correlated features
x = x.drop(['JobLevel','PerformanceRating','YearsInCurrentRole','YearsWithCurrManager'],axis=1)
```

```python
x.shape
```

Now there are only 25 features in dataset

### v) Standard Scaling

```python
# Importing standardScaler and scaling the data
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(x)
X=pd.DataFrame(X, columns=x.columns)
X.describe()
```

The entire data is scaled as the unit standard deviation for all the features. So the modeling will take less computational cost.

## MODELING

```python
# To split the data to training and validation 
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state =123, test_size = .2)
```

x and y are split by 80:20 ratio for training and testing respectively.

### i) Linear Regression

```python
#importing linear regression from linear model
from sklearn.linear_model import LinearRegression
l_model = LinearRegression()
l_model.fit(X_train,y_train)
y_predict = l_model.predict(X_test)
```

l_model variable is trained by using X_train & y_train and predicted y values Corresponding to X_test

```python
# Importing RMSE, R Square functions to cross validate the predicted y values and actual y_test values 
from sklearn.metrics import mean_squared_error,r2_score
```

```python
print("RMSE is : ",mean_squared_error(y_test,y_predict,squared=False))
print("R squared value is : ",r2_score(y_test,y_predict))
```

RMSE and  R Squared values are calculated to cross validate the predicted y values and actual y_test values. RMSE is :  3014 (High value) and R squared value is :  0.6385 (Low value). i.e., Linear Regression is NOT a good model for this dataset.

### ii) KNearest Neighbor Regressor

```python
# Importing KNearest Neighbor regressor
from sklearn.neighbors import KNeighborsRegressor
```

```python
#creating empty list for finding optimal K values by checking R2 values
r2_value = []
#creating neighbors as 3 to 15
neighbors = np.arange(3,15)
#By using a for loop each neighbors' y predict values are generated & cross validated with 'y_test' with 'R squared value'.
for k in neighbors:
    # taking instances for KNN into variable name regr using the method minkowski
    regr = KNeighborsRegressor(n_neighbors=k, metric="minkowski")
    regr.fit(X_train,y_train)
    y_pred = regr.predict(X_test)
    r2 = r2_score(y_test,y_pred)
#     Each R2 value is appended to r2_value list
    r2_value.append(r2)
```

```python
r2_value
```

By checking the above R squared values (all are lessthan .5), we can clearly say that KNearest Neighbor regressor is not good for this dataset.
